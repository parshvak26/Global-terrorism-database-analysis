````python
#function to make a column of decade by taking value of year
def decade(x):
  x=int(x)
  if 1970<= x <= 1979:
    return "70s"
  elif 1980 <= x <=1989:
    return "80s"
  elif 1990 <= x <=1999:
    return "90s"
  elif 2000 <= x <= 2009:
    return "00s"
  elif 2010 <= x <= 2020:
    return "10s"
  else:
    return 'not define'
````

````python
#Applying decade functio here to get the column in which decade of attack is shown
df_time['decade'] = df_time.apply(lambda row: decade(row['year']),axis=1)
````
